**Przypomnienie**:

**Autokowariancja**
\begin{equation*}
    \hat{\gamma}(h) = \frac{1}{n} \sum_{i=1}^{n-h}(y_i - \overline{y})(y_{i+h} - \overline{y})
\end{equation*}

**Autokorelacja**
\begin{equation*}
    \text{ACF}(h) = \frac{\hat{\gamma}(h)}{\hat{\gamma}(0)}
\end{equation*}
